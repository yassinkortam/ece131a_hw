```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
```

# Tossing a Fair and Unfair Die
### a)

```python
#Randomly sample a discrete uniform random variable n times 
sample_sizes = reversed([5E1, 1E2, 1E3, 2E3, 3E2, 1E4, 1E5])
p_odd = {"Sample size":[],"Frequency":[], "Probability":[]}
plt.figure(figsize=(4, 3))
plt.title("Histogram of a Discrete Uniform Random Variable")
plt.xlabel("Value")
plt.ylabel("Frequency")
for sample_size in sample_sizes:
    x = np.random.randint(1, 11, size=int(sample_size))
    p_odd["Sample size"].append(sample_size)
    p_odd["Frequency"].append(np.count_nonzero(x % 2))
    p_odd["Probability"].append(np.count_nonzero(x % 2) / sample_size)
    plt.hist(x, label="n = " + str(sample_size))
plt.legend(loc="upper right")
plt.show()

p_odd = pd.DataFrame(p_odd)
p_odd
```

- The expiremental probability of sampling an odd number from the die is about 0.5, which is similar to the theoretical value.

### b)

### c)

### d)
- The probability of sampling an odd number from a 10 sided die, where the probability of sampling an odd number is double that of sampling an even number and even numbers are evenly distributed, is theoretically about 0.667. 

```python
#Biased 10 sided die that is twice as likely to pick an odd number
def biased_die():
    odd = np.random.randint(1, 4, size=1)
    if odd % 2 == 0:
        x = np.random.randint(1, 11, size=1)
        while x % 2 != 0:
            x = np.random.randint(1, 11, size=1)
        return x
    else:
        x = np.random.randint(1, 11, size=1)
        while x % 2 == 0:
            x = np.random.randint(1, 11, size=1)
        return x

#Randomly sample a biased die n times
sample_sizes = reversed([5E1, 1E2, 1E3, 2E3, 3E2, 1E4, 1E5])
p_odd_biased = {"Sample size":[],"Frequency":[], "Probability":[]}
plt.figure(figsize=(4, 3))
plt.title("Histogram of a Biased Die")
plt.xlabel("Value")
plt.ylabel("Frequency")
for sample_size in sample_sizes:
    x = np.array([biased_die() for i in range(int(sample_size))])
    p_odd_biased["Sample size"].append(sample_size)
    p_odd_biased["Frequency"].append(np.count_nonzero(x % 2))
    p_odd_biased["Probability"].append(np.count_nonzero(x % 2) / sample_size)
    plt.hist(x, label="n = " + str(sample_size))
plt.legend(loc="upper right")
plt.show()

p_odd_biased = pd.DataFrame(p_odd_biased)
p_odd_biased
```

- The expiremental probability of sampling an odd number from the die is about 0.667, which is similar to the theoretical value.

# Maximum Likelihood Estimation

### a) 
#### Proof
$$
{f_{x_{1},...x_{n}}\left(x_{1},...x_{n}|\mu,\sigma\right)=\prod_{i=1}^{n}\frac{1}{{\sqrt[]{2\pi}}\sigma}e^{\frac{\left(x_{i}-\mu\right)^{2}}{2\sigma^{2}}}}\\
$$
$$
\ln\left|f_{x_{1},...x_{n}}\left(x_{1},...x_{n}|\mu,\sigma\right)\right|=\sum_{i=1}^{n}\ln\left|\frac{1}{{\sqrt[]{2\pi}}\sigma}\right|-\frac{\left(x_{i}-\mu\right)^{2}}{2\sigma^{2}}=\frac{-n}{2}\ln\left|\frac{1}{{\sqrt[]{2\pi}}\sigma}\right|-\sum_{i=1}^{n}\frac{\left(x_{i}-\mu\right)^{2}}{2\sigma^{2}}
$$
$$
{Q.E.D.}\\
$$

### b)
#### Mean Maximum Likelihood Estimation
$$
{\mu_{MLE}=\arg \max_{\mu}\left[f_{x_{1},...x_{n}}\left(x_{1},...x_{n}|\mu,\sigma\right)\right]}\\
$$
$$
\frac{d}{d\mu}\left[\frac{-n}{2}\ln\left|2\pi \sigma^{2}\right|-\sum_{i=1}^{n}\frac{\left(x_{i}-\mu\right)^{2}}{2\sigma^{2}}\right]|_{\mu=\mu_{MLE}} = \left[\frac{-n\mu}{\sigma^{2}}+\frac{1}{\sigma^{2}}\sum_{i=1}^{n}x_{i}\right]|_{\mu=\mu_{MLE}} = 0
$$
$$
{\mu_{MLE}=\frac{1}{n}\sum_{i=1}^{n}x_{i}}	
$$
#### Standard Deviation Maximum Likelihood Estimation
$$
{\sigma_{MLE}=\arg \max_{\sigma}\left[f_{x_{1},...x_{n}}\left(x_{1},...x_{n}|\mu,\sigma\right)\right]}\\
$$
$$
\frac{d}{d\sigma}\left[\frac{-n}{2}\ln\left|2\pi \sigma^{2}\right|-\sum_{i=1}^{n}\frac{\left(x_{i}-\mu\right)^{2}}{2\sigma^{2}}\right]|_{\sigma=\sigma_{MLE}}^{}=\left[\frac{-n}{\sigma}+\frac{1}{\sigma^{3}}\sum_{i=1}^{n}\left(x_{i}-\mu\right)^{2}\right]|_{\sigma=\sigma_{MLE}}^{}=0
$$
$$
\sigma_{MLE}={\sqrt[]{\frac{1}{n}\sum_{i=1}^{n}\left(x_{i}-\mu\right)^{2}}}
$$

### c)

```python
#import data in data.txt into a pandas dataframe
data = pd.read_csv("data.txt", header=None)
mean = sum(data[0])/len(data[0])
std = np.sqrt(sum(abs(data[0]-mean)**2)/len(data[0]))

print("Mean: " + str(mean))
print("Standard Deviation: " + str(std))

gaussian = norm(mean, std)

#plot the pdf
plt.figure(figsize=(4, 3))
plt.title("PDF of Data")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.hist(data, bins=200, density=True)
x = np.linspace(-20, 60, 100)
plt.plot(x, gaussian.pdf(x), label='Gaussian PDF')
plt.legend(loc='upper right')
plt.show()
```

# Naive Bayes Classifier

### a)

```python
# Load data from csv file
data = pd.read_csv("user_data.csv")

pmf_B = data["Bought"].value_counts(normalize=True)
pmf_T = data["Spender Type"].value_counts(normalize=True)
pmf_S = data["Sex"].value_counts(normalize=True)
pmf_A = data["Age"].value_counts(normalize=True)

#Plot the probability distributions of the data
fig = plt.figure()

fig.add_subplot(2, 2, 1)
plt.title("PMF of B")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.bar(pmf_B.index, pmf_B.values, label="B")
plt.legend(loc="upper right")

fig.add_subplot(2, 2, 2)
plt.title("PMF of T")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.bar(pmf_T.index, pmf_T.values, label="T")
plt.legend(loc="upper right")

fig.add_subplot(2, 2, 3)
plt.title("PMF of S")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.bar(pmf_S.index, pmf_S.values, label="S")
plt.legend(loc="upper right")

fig.add_subplot(2, 2, 4)
plt.title("PMF of A")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.bar(pmf_A.index, pmf_A.values, label="A")
plt.legend(loc="upper right")

plt.tight_layout()
plt.show()
```

### b)

```python
#Find conditional PMFs of T, S, and A given B=0 and B=1
pmf_T_given_B0 = data[data["Bought"] == 0]["Spender Type"].value_counts(normalize=True)
pmf_S_given_B0 = data[data["Bought"] == 0]["Sex"].value_counts(normalize=True)
pmf_A_given_B0 = data[data["Bought"] == 0]["Age"].value_counts(normalize=True)
pmf_T_given_B1 = data[data["Bought"] == 1]["Spender Type"].value_counts(normalize=True)
pmf_S_given_B1 = data[data["Bought"] == 1]["Sex"].value_counts(normalize=True)
pmf_A_given_B1 = data[data["Bought"] == 1]["Age"].value_counts(normalize=True)

#Plot the conditional PMFs
fig = plt.figure(figsize=(8, 6))

fig.add_subplot(2, 3, 1)
plt.title("PMF of T given B=0")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.bar(pmf_T_given_B0.index, pmf_T_given_B0.values, label="T")
plt.legend(loc="upper right")

fig.add_subplot(2, 3, 2)
plt.title("PMF of S given B=0")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.bar(pmf_S_given_B0.index, pmf_S_given_B0.values, label="S")
plt.legend(loc="upper right")

fig.add_subplot(2, 3, 3)
plt.title("PMF of A given B=0")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.bar(pmf_A_given_B0.index, pmf_A_given_B0.values, label="A")
plt.legend(loc="upper right")

fig.add_subplot(2, 3, 4)
plt.title("PMF of T given B=1")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.bar(pmf_T_given_B1.index, pmf_T_given_B1.values, label="T")
plt.legend(loc="upper right")

fig.add_subplot(2, 3, 5)
plt.title("PMF of S given B=1")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.bar(pmf_S_given_B1.index, pmf_S_given_B1.values, label="S")
plt.legend(loc="upper right")

fig.add_subplot(2, 3, 6)
plt.title("PMF of A given B=1")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.bar(pmf_A_given_B1.index, pmf_A_given_B1.values, label="A")
plt.legend(loc="upper right")

plt.tight_layout()
plt.show()
```

### c)
#### Given Conditional Independence
$$
{P\left[B,T,S,A\right]=P\left[T,S,A|B\right]\cdot P\left[B\right]}\\
{P\left[B,T,S,A\right]=P\left[T|B\right]\cdot P\left[S|B\right]\cdot P\left[A|B\right]\cdot P\left[B\right]}\\
$$

```python
def P(B, T, S, A):
    if B == 0:
        return pmf_T_given_B0[T] * pmf_S_given_B0[S] * pmf_A_given_B0[A] * pmf_B[0]
    if B == 1:
        return pmf_T_given_B1[T] * pmf_S_given_B1[S] * pmf_A_given_B1[A] * pmf_B[1]
    else:
        return -1

p_B0_T1_S0_A55 = P(0, 1, 0, 55)
p_B1_T1_S0_A55 = P(1, 1, 0, 55)

print("P(B=0,T=1,S=0,A<=55) = " + str(p_B0_T1_S0_A55))
print("P(B=1,T=1,S=0,A<=55) = " + str(p_B1_T1_S0_A55))
```

### d)
#### Applying Bayes' Rule
$$
P\left[B_{j}|T,S,A\right]=\frac{P\left[B_{j},T,S,A\right]}{\sum_{k=1}^{n}P\left[T|B_{k}\right]\cdot P\left[S|B_{k}\right]\cdot P\left[A|B_{k}\right]\cdot P\left[B_{k}\right]}
$$

```python
#Condition B on T, S, A
def P_conditional_B(B, T, S, A):
    return P(B, T, S, A) / ((pmf_T_given_B0[T] * pmf_S_given_B0[S] * pmf_A_given_B0[A])*pmf_B[0] + (pmf_T_given_B1[T] * pmf_S_given_B1[S] * pmf_A_given_B1[A])*pmf_B[1])

p_B0_conditional_T1_S0_A55 = P_conditional_B(0, 1, 0, 55)
p_B1_conditional_T1_S0_A55 = P_conditional_B(1, 1, 0, 55)

print("P(B=0|T=1,S=0,A<=55) = " + str(p_B0_conditional_T1_S0_A55))
print("P(B=1|T=1,S=0,A<=55) = " + str(p_B1_conditional_T1_S0_A55))
```

- The probability that a female below the age of 55 who is a large spender to buy a product is ~0.33

# Central Limit Theorem

### a)

```python
#Number of samples from Zn
t = 1E4

#Function to generate Zn
def Z(n):
    return sum(np.random.uniform(3, 7, size=n))/n

# Plot the pdf of Zn
fig = plt.figure(figsize=(8, 6))
for count, n in enumerate([1, 2, 3, 10, 30, 100]):
    z = np.array([Z(n) for i in range(int(t))])
    fig.add_subplot(2, 3, count+1)
    plt.title("PDF of Z" + str(n))
    plt.xlabel("Value")
    plt.ylabel("Probability Density")
    plt.hist(z, density=True, label="Z" + str(n))
    plt.legend(loc="upper right")
plt.tight_layout()
plt.show()
```

### b)

Zn is the sample mean of Xi ~uniform(3,7)

- The mean of Zn is simply the mean of the sample mean
E[Zn] ~ E[Xi] = 5

- The variance of Zn is simply the variance of the sample mean 
VAR[Zn] ~ VAR[Xi]/n = 25/12n

### c)

```python
# Plot the pdf of Zn with a normal distribution superimposed
fig = plt.figure(figsize=(8, 6))
for count, n in enumerate([1, 2, 3, 10, 30, 100]):
    z = np.array([Z(n) for i in range(int(t))])
    gaussian = norm(5, 2.5/np.sqrt(3*n))
    fig.add_subplot(2, 3, count+1)
    plt.title("PDF of Z" + str(n))
    plt.xlabel("Value")
    plt.ylabel("Probability Density")
    plt.hist(z, density=True, label="Z" + str(n))
    plt.plot(np.linspace(0, 9, 100), gaussian.pdf(np.linspace(0, 9, 100)), label="Normal")
    plt.legend(loc= "upper right")
plt.tight_layout()
plt.show()
```

### d)

Z2n is the sample mean of Xi ~uniform[1,10]

- The mean of Zn is simply the mean of the sample mean
E[Zn] ~ E[Xi] = 5.5

- The variance of Zn is simply the variance of the sample mean 
VAR[Zn] ~ VAR[Xi]/n = 100/12n

```python
def Z2n(n):
    return sum(np.random.randint(1, 11, size=n))/n

# Plot the pdf of Z2n with a normal distribution superimposed
fig = plt.figure(figsize=(8, 6))
for count, n in enumerate([1, 2, 3, 10, 30, 100]):
    z = np.array([Z2n(n) for i in range(int(t))])
    gaussian = norm(5.5, 5/np.sqrt(3*n))
    fig.add_subplot(2, 3, count+1)
    plt.title("PDF of Z" + str(n))
    plt.xlabel("Value")
    plt.ylabel("Probability Density")
    plt.hist(z, density=True, label="Z" + str(n))
    plt.plot(np.linspace(0, 9, 100), gaussian.pdf(np.linspace(0, 9, 100)), label="Normal")
    plt.legend(loc="upper right")
plt.tight_layout()
plt.show()
```

### e)

The PDF of Zn is simply the PDF of some uniform random variable Yj within the interval a/n <= y <= b/n convolved with itself n times
- The sum of independent random variables is their convolution.
- Dividing a uniform random variable by a number n changes its bounds.

```python
# Plot the exact pdf of Zn with a normal distribution superimposed
def Zn_e(n):
    # pdf of (1/n)xj aka yj
    final_length = 10
    t = n
    a = 3
    b = 7
    yj = np.zeros(int(t*b/n) + int(t*a/n))
    yj[int(t*a/n)+1:int(t*b/n)] = 1
    zn = yj
    for n in np.arange(1, n):
        zn = np.convolve(zn, yj)
    
    #normalize zn
    zn = zn/sum(zn)

    #remove near zero values
    zn = zn[zn > 1E-10]
    return zn

fig = plt.figure(figsize=(8, 6))
for count, n in enumerate([1, 2, 3, 10, 30, 100]):
    zn = Zn_e(n)
    fig.add_subplot(2, 3, count+1)
    plt.title("PDF of Z" + str(n))
    plt.xlabel("Value")
    plt.ylabel("Probability Density")
    plt.plot(np.linspace(0, 9, num=len(zn)), zn, label="Z" + str(n))
    plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
