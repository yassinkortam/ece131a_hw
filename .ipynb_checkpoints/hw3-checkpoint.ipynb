```python
# Yassin Kortam
# EC ENGR 131A
# Problem 5

import matplotlib.pyplot as plt 
import numpy as np
from math import comb

#Uniform pmf
def pmfA(k):
    return 0.25

#Binomial pmf
def pmfB(k):
    return comb(5, k) * (0.25**k) * ((1-0.25)**(5-k))

#Geometric pmf
def pmfC(k):
    return 0.5*(1-0.5)**k
```

```python
#Generate series
pmf_a = np.array([pmfA(k) for k in range(4)])
pmf_b = np.array([pmfB(k) for k in range(6)])
pmf_c = np.array([pmfC(k) for k in range(6)])

#Plot pmf
plt.figure()
plt.title("A: Uniform PMF")
plt.stem(pmf_a)
plt.figure()
plt.title("B: Binomial PMF")
plt.stem(pmf_b)
plt.figure()
plt.title("C: Geometric PMF")
plt.stem(pmf_c)
```
