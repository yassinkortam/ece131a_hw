```python
import numpy as  np
import matplotlib.pyplot as plt

def Y(x, a, b):
    return a*x + b

def Z(x):
    return np.power(x, 2)

def W(x):
    return np.power(x, 3)
```

### b) 
X and Y are perfectly linearly correlated, hence p = 1

```python
#Randomly sample 1000 points from the standard gaussian (a=5, b=0)
plt.figure()
plt.title("X vs Y(x, 5, 0)")
x = np.random.normal(0, 1, 1000)
y = Y(x, 5, 0)
plt.scatter(x, y)
```

### c)
X and Y are perfectly linearly correlated, hence p = 1

```python
#Randomly sample 1000 points from the standard gaussian (a=-4, b=0)
plt.figure()
plt.title("X vs Y(x, -4, 0)")
x = np.random.normal(0, 1, 1000)
y = Y(x, -4, 0)
plt.scatter(x, y)
```

### d)
X and Y are quadratically correlated (non-linear) hence p = 0

```python
#Randomly sample 1000 points from the standard gaussian and get Z
plt.figure()
plt.title("X vs Z(x)")
x = np.random.normal(0, 1, 1000)
z = Z(x)
plt.scatter(x, z)
```

### e)
X and Y are cubically correlated (non-linear) hence p = 0

```python
#Randomly sample 1000 points from the standard gaussian and get W
plt.figure()
plt.title("X vs W(x)")
x = np.random.normal(0, 1, 1000)
w = W(x)
plt.scatter(x, w)
```
