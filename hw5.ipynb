```python
import numpy as np
import matplotlib.pyplot as plt

SIZE = 5
STEP = 0.1
P = 0.6 
L = 1

def M(k, p):
    return (1-p)**k

def X(l, t):
    return np.exp(-1*l*t)
```

```python
k = np.arange(0, SIZE)
t = np.arange(0, SIZE, STEP)
Mk = list(map(M, k, np.full(len(k), P)))
Xt = list(map(X, t, np.full(len(t), L)))

plt.figure()
plt.plot(k, Mk, label="Geometric")
plt.plot(t, Xt, label="Exponential")
plt.legend()
plt.show()
```
