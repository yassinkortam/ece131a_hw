```python
# Yassin Kortam
# EC ENGR 131A

import matplotlib.pyplot as plt 
import numpy as np
from math import comb

#Constants
RANGE = 25
RES = 0.1

#Problem 3
def cdf(k,b):
    if k < 0:
        return 0
    return 1 - np.exp(-1*k**b)

```

```python
#Generate series
for b in [0.5, 1, 2]:
    my_cdf = np.array([cdf(k, b) for k in np.arange(-1*RANGE,RANGE,RES)])
    x = np.arange(-1*RANGE,RANGE,RES)

    #Plot pmf
    plt.figure()
    plt.title('CDF for b = {b}'.format(b = b))
    plt.plot(x, my_cdf)
```

```python
x = np.log(np.arange(0, RANGE, RES))
y = -1*np.arange(0,RANGE,RES)**2

#Plot pmf
plt.figure()
plt.title('lnP[X > x] vs lnx')
plt.ylabel('lnP[X > x]')
plt.xlabel('lnx')
plt.plot(x, y)
```
