# HW6 Problem 5
(a) Generate 5000 samples of independent zero-mean, unit-variance Gaussian random variables X1 and X2 by the Box-Muller method (i.e by generating 5000 samples of U1 and U2 each and then using the transformation to get 5000 samples each of X1 and X2).
<br><br>
(b) Plot the histogram of 5000 samples of X1 and compare it with the pdf of a zeromean unit variance Gaussian random variable. Repeat the same for X2.
<br><br>
(c) Plot the samples of X1 vs X2 in a scatter plot. Does the plot have a circular symmetry? Explain.

## a)

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss

#Box-Muller Gaussian Random Variable Generator
def X1():
    return np.sqrt(-2*np.log(np.random.uniform()))*np.cos(2*np.pi*np.random.uniform())
def X2():
    return np.sqrt(-2*np.log(np.random.uniform()))*np.sin(2*np.pi*np.random.uniform())
```

## b)

```python
#Box-Muller Gaussian Random Variables
x1 = np.array([X1() for i in np.arange(5000)])
x2 = np.array([X2() for i in np.arange(5000)])

#Standard Gaussian PDF
x = np.arange(-5,5,0.01)
gx = ss.norm.pdf(x)

plt.figure()
plt.title("X1 & Standard Gaussian")
plt.hist(x1, density=True)
plt.plot(x, gx, color="orange")
plt.show()

plt.figure()
plt.title("X2 & Standard Gaussian")
plt.hist(x2, density=True)
plt.plot(x, gx, color="orange")
plt.show()
```

The histograms of both X1 and X2 resemble an interpolation of a low rate sampling of a normal distribution.

## c)

```python
fig, ax = plt.subplots()
plt.title("X1 vs X2")
plt.scatter(x1, x2)
ax.add_patch(plt.Circle((0,0), radius=3, fill=False))
plt.show()
```

The scatter plot X1 vs X2 is circular because the Box Muller method equations resemble the parametric equations for a circle, where the radius is determined by the positive half of a bell curve. I plotted a circle with a radius of 3 standard deviations, which should contain 99.7% of all points.
